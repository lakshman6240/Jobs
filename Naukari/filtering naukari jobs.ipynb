```python
spark.sql('truncate table staging_catalogue.jobs.naukarijobs')
all_jobs = spark.table("raw_catalogue.jobs.naukarijobs")
jobs = all_jobs.select("jdURL","jobDescription","tagsAndSkills").collect()
```

```python
scored_jobs = score_job(jobs, "jdURL", "jobDescription", "tagsAndSkills")
```

```python
scored_df = spark.createDataFrame(scored_jobs)
df = scored_df.join(all_jobs, scored_df.url==all_jobs.jdURL)
filtered_df = df.filter(df.score > 60)
filtered_df.write.mode('overwrite').option("mergeSchema", "true").saveAsTable("staging_catalogue.jobs.naukarijobs")
```
